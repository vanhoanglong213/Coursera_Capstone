```python
import pandas as pd
```

```python
data_table = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

```python
len(data_table)
```

```python
data_table[0].info()
```

```python
df = pd.DataFrame(data_table[0])
```

```python
df
```

```python
df = df[df.Borough != 'Not assigned']
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
neighborhoodxs = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(lambda tags: ','.join(tags)).reset_index()
```

```python
neighborhoodxs
```
